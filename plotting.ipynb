```python
import math
import numpy as np

def qiur(x,y):
    r=np.sqrt(x*x+y*y)
    return r
def qiuQ(x,y):
    Q=np.arctan(y/x)*180/math.pi
    return Q
a_x=[]
a_y=[]
for x,y in zip(high_x,high_y):
    a_x.append(qiur(x,y))
    a_y.append(qiuQ(x,y))

```

```python
plt.figure(figsize=(20, 10))
for x, y in zip(a_x, a_y):
    plt.polar(x, y, 'ro')
plt.show()
```
