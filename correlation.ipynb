```python
def trans_data(data):
  label=data["labels"]
  M_fnirs=data["M_fnirs"]
  M_eeg=data["M_eeg"]
  time_fnirs=data["time_fnirs"]
  time_eeg=data["time_eeg"]
  return label,M_fnirs,M_eeg,time_fnirs,time_eeg
```

```python
path = (r'../EEG-fNIRs/data/')
fname = "M_all_Subind_31_32_33.mat"
fname1= "M_all_Subind_41_32_33.mat"
fname2= "M_all_Subind_91_32_33.mat"
fname3= "M_all_Subind_101_32_33.mat"

import os
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
data = sio.loadmat(path+fname)
data1=sio.loadmat(path+fname1)
data2=sio.loadmat(path+fname2)
data3=sio.loadmat(path+fname3)
```

```python
label,M_fnirs,M_eeg,time_fnirs,time_eeg=trans_data(data)
label1,M_fnirs1,M_eeg1,time_fnirs1,time_eeg1=trans_data(data1)
label2,M_fnirs2,M_eeg2,time_fnirs2,time_eeg2=trans_data(data2)
label3,M_fnirs3,M_eeg3,time_fnirs3,time_eeg3=trans_data(data3)
```

```python
M_eeg=M_eeg[:,:,:,20:100]
M_eeg1=M_eeg1[:,:,:,20:100]
M_eeg2=M_eeg2[:,:,:,20:100]
M_eeg3=M_eeg3[:,:,:,20:100]
```

```python
def calculate_corr_tartar_tarnontar(target,nontarget,target_ref,nontarget_ref):
  corr_tar_tar=[];
  corr_tar_nontar=[];
  [length,width]=target.shape
  for i in range(length):
    corr_tar_tar.append((np.corrcoef(target[i],target_ref[i]))[1][0])
    corr_tar_nontar.append((np.corrcoef(target[i],nontarget[i]))[1][0])
  return corr_tar_tar,corr_tar_nontar
```

```python
def aver_target_non_target(signal,lab):
  target=[] #0 represent target
  non_target=[] #1 represent non-target
  [a,b,c,d]=signal.shape
  for i in range(a):
    if (lab[i] == 1):  #target
      target.append(signal[i])
    if (lab[i] == 0): #non-target
      non_target.append(signal[i])
  av_target=sum(target)/len(target);
  av_non_target=sum(non_target)/len(target);
  av_target=np.reshape(av_target,(c,d))
  av_non_target=np.reshape(av_non_target,(c,d))
  return av_target, av_non_target
```

```python
target_eg,nontarget_eg=aver_target_non_target(M_eeg,label)
target_eg_ref,nontarget_eg_ref=aver_target_non_target(M_eeg1,label1)
target_eg_tp2,nontarget_eg_tp2=aver_target_non_target(M_eeg2,label2)
target_eg_ref_tp2,nontarget_eg_ref_tp2=aver_target_non_target(M_eeg3,label3)
```

```python
corr_tar_tar_eg,corr_tar_nontar_eg=calculate_corr_tartar_tarnontar(target_eg,nontarget_eg,target_eg_ref,nontarget_eg_ref)
corr_tar_tar_eg_tp2,corr_tar_nontar_eg_tp2=calculate_corr_tartar_tarnontar(target_eg_tp2,nontarget_eg_tp2,target_eg_ref_tp2,nontarget_eg_ref_tp2)
```

```python
import pandas as pd
df=pd.DataFrame(data=corr_tar_tar_eg,columns=['target-target-EEG-large'],index=['A','B','P5','P4','Cp2','Cp1','C1/3','Cz','FCz','C2/4','FC6','F4','F3','FC5','H','G'])
```
